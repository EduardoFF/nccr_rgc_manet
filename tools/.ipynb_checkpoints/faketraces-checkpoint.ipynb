```python
timestep=1
speed=0.5
x1=487200
x2=487500
y1=113200
y2=113700
simlen=3600
```

```python
def randomPoint():
    import random
    x=random.uniform(x1,x2)
    y=random.uniform(y1,y2) 
    return round(x),round(y)
def dist((px,py),(qx,qy)):
    return ((px-qx)**2 + (py-qy)**2)**0.5
```

```python
import math
pts=[randomPoint()]
t=0
while t<simlen:
    (cx,cy) = pts[-1]
    (nx,ny) = randomPoint()
    oi = math.atan2(ny-cy,nx-cx)
    di = timestep*speed
    while t<simlen:
        if dist((cx,cy),(nx,ny)) < di:
            break
        cx += round(di*math.cos(oi))
        cy += round(di*math.sin(oi))
        pts+=[(cx,cy)]
        t+=1
```
